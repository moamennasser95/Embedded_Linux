```python
cap = cv2.VideoCapture(0)

cnt = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if(ret):
        cnt +=1
        cv2.imshow('Frame',frame)
        k = cv2.waitKey(1)

        if k == ord('q'):
            break
        elif k == ord('s'):
            cv2.imwrite('Files/out_{}.jpg'.format(cnt),frame)

    else:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def nothing(x):
    pass

cv2.namedWindow('Image')

cv2.createTrackbar('val','Image',0,255,nothing)

img = np.zeros((500,500,1),np.uint8)

while 1:
    val = cv2.getTrackbarPos('val','Image')
    img[:] = val
    cv2.imshow('Image',img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```
